```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Function to detect outliers using IQR
outliers = {}
for column in numeric_df.columns:
    Q1 = numeric_df[column].quantile(0.25)
    Q3 = numeric_df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers[column] = numeric_df[(numeric_df[column] < lower_bound) | (numeric_df[column] > upper_bound)][column]

# Print outliers
for column, outlier_values in outliers.items():
    print(f"Outliers in {column}:")
#     print(outlier_values.values)
    print(f"Number of outliers: {len(outlier_values)}")
    print("---")

print("Outlier detection completed.")
# outliers
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the number of numeric columns
num_cols = len(numeric_df.columns)

# Calculate the number of rows and columns for the subplots
num_rows = (num_cols + 1) // 2  # Round up to the nearest integer
num_cols = 2

# Create box plots for numeric columns to visualize outliers
plt.figure(figsize=(15, 5 * num_rows))
for i, column in enumerate(numeric_df.columns, 1):
    plt.subplot(num_rows, num_cols, i)
    sns.boxplot(y=numeric_df[column])
    plt.title(f'Box Plot of {column}')
    plt.tight_layout()
# plt.savefig('outliers_box_plots.png')
# plt.close()

print("Box plots for outlier visualization completed.")

# Display summary statistics
# print("\
# Summary statistics:")
# print(numeric_df.describe())
```

```python
# Histograms
plt.figure(figsize=(15, 10))
numeric_df.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
# plt.savefig('histograms.png')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True,cmap='magma', linewidths=0.5)
plt.title('Correlation Heatmap')
# plt.savefig('correlation_heatmap.png')
plt.show()

print("Histograms and correlation heatmap have been generated.")
```

```python
sales_channel_performance = df['Sales Channel'].value_counts()
# 1. Sales Channel Performance
plt.figure(figsize=(10, 6))
sales_channel_performance.plot(kind='bar', color='c')
plt.title('Sales Channel Performance')
plt.xlabel('Sales Channel')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
# plt.savefig('sales_channel_performance.png')
plt.show()
```

```python
# 4. Seasonal Trends
monthly_sales_trend = df.groupby(df['OrderDate'].dt.to_period('M'))['Unit Price'].sum().reset_index()
monthly_sales_trend['OrderDate'] = monthly_sales_trend['OrderDate'].dt.to_timestamp()
# 4. Monthly Sales Trend
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales_trend['OrderDate'], monthly_sales_trend['Unit Price'], marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
# plt.savefig('monthly_sales_trend_v2.png')
plt.show()
```
